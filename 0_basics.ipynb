# Cuerpos
### Declarando un cuerpo

```python
F = GF(31)
F(5) + F(5)   # Resultado F(10)
```

```python
F(31) + F(5)  # Resultado F(5)
```

### Ejercicios

```python
# Ejercicio 1: Calcula F(7) + F(8) + F(15) en F_31
# ¿Cuál debería ser el resultado?

# Tengo que hacer 7 + 8 + 15 mod 31. Es decir 30 mod 31. O sea: 30.
# F ya es GF(31)
F(7) + F(8) + F(15)
```

```python
# Ejercicio 2: Calcula F(5) * F(6) y F(20) * F(3) en F_31
# ¿Son iguales?

# En un cuerpo si yo multiplico y luego aplico modulo la relación respecto al cuerpo se mantiene
# Entonces si hago 5*6 en los naturales obtengo 30. En F_31 es 30 mod 31 = 30.
# Y 20*3 en los naturales es 60. En F_31 es 60 mod 31 = 29.
# No son iguales

print(F(5) * F(6))
print(F(20) * F(3))
```

```python
# Ejercicio 3: Encuentra el inverso multiplicativo de F(3) en F_31
# Buscamos x tal que x * x^(-1) = F(1)

# No tengo presente ahora si hay una forma de hacerlo matematicamente
# Si se que dado un cuerpo GF, y un elemento de ese cuerpo, debe existir un inverso multiplicativo (y debe estar dentro del cuerpo).
# Entonces se que puedo recorrer todos los numeros del cuerpo hasta encontrarlo
for elem in F:
    if elem * F(3) == F(1):
        print(elem)

# El 21 es el inverso multiplicativo de 3 en F_31
```

```python
# Ejercicio 4: Encuentra un generador para F_31
# Un generador g es un elemento donde g^i mod 31 nos da todos los elementos distintos de cero en F_31

# Los candidatos a ser generador son cada uno de los elementos del cuerpo. Entonces, para cada elemento del cuerpo tengo que
# fijarme si ese elemento es capaz de generar todos los elementos distintos de cero en F_31

for gen in F:
    elems_generated = set([gen**i for i in F])
    if len(elems_generated) == 30:
        print(f"{gen} es un generador del cuerpo")

# Tengo varios generadores. 3, 11, 12, 13, 17, 21, 22, 24.
```
